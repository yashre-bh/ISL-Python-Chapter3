### (a)

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
multireg = smf.ols('Sales~Urban + Price + US', data=car).fit()
```

```python
multireg.summary()
```

p-value for Urban is very close to 1, hence we can accept null hypothesis for X='Urban'
Null hypothesis accordingly rejected for Price and US
### (b)

1. Negative relation between (Price and sales), (Urban and sales) while positive relation between (US and sales). Although as concluded above, we can safely conclude that 'Urban' does not significantly affect the Sales.


### (c)
1. sales = 13.0435 - 0.0219(Urban) + 1.2006(US) - 0.0545(Price), where Urban and US can hold only values 0 and 1, 1 for yes and 0 for no. 

### (d)
As concluded above, Null hypothesis can be rejected for Price and US

### (e)

```python
multireg2 = smf.ols('Sales~Price+US', data=car).fit()
multireg2.summary()
```

### (f)
Both models have same R^2 values, and even the adjusted R^2 values are almost equal. Hence both the models fit almost equally

### (g)
Confidence intervals:

```python
print('For (a): \n', multireg.conf_int())
print('For (e): \n', multireg2.conf_int())
```
